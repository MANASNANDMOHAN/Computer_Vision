```python
#A)Open the Image
from PIL import Image
img=Image.open('elon.jpg')
```

```python
#B)Show the Image
display(img)
```

```python
#C)Display the basic Image properties
print('Format / Type of the image:' ,img.format)
print(img.mode)
print(img.size)
print(img.width)
print(img.height)
```

```python
#D)Using pillow load an Image & Convert it into an array.
#Also display the properties of image array
from PIL import Image
from numpy import asarray
import numpy
#load the image
img = Image.open("elon.jpg")
#Convert image to numpy array
data = asarray(img)
print(data)
print(len(data))
print(data.shape)
```

```python
#E)Using pillow load the Image & convert it into numpy array 
#Report shape of the array
#Finally Convert the array back into Pillow Image
#Print Details(format,mode,size,width,height)
from PIL import Image
from numpy import asarray
import numpy
#load the image
img = Image.open('elon.jpg')
data=asarray(img)
print(data)
print(data.shape)
print(data.size)
data2=Image.fromarray(data)
print(data2)
```

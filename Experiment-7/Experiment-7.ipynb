```python
import cv2
import matplotlib.pyplot as plt
img_1=cv2.imread('jim.png')
img_2=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
img_3=cv2.cvtColor(img_1,cv2.COLOR_BGR2GRAY)
img_4=cv2.cvtColor(img_1,cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.imshow(img_1)
plt.title('BGR')
plt.axis('off')
plt.subplot(1,4,2)
plt.imshow(img_2)
plt.title('RGB')
plt.axis('off')
plt.tight_layout()
plt.subplot(1,4,3)
plt.imshow(img_3)
plt.title('BGR2GRAY')
plt.axis('off')
plt.subplot(1,4,4)
plt.imshow(img_1)
plt.title('RGB2GRAY')
plt.axis('off')
```
